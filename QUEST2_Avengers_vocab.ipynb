```python
import re
from collections import Counter

def encoder():

  with open('Avengers.txt',"r") as f:
    data = f.read().splitlines()

    # 소문자로 변경
    data = [x.lower() for x in data]
    # print(data)

    # 기호를 제거
    data = [re.sub(r"[^가-힣a-zA-Z0-9' ]","",x) for x in data]
    # print(data)

    # 단어 단위로 분리
    data = " ".join(data)
    data = data.split()

    # 단어 빈도 수 체크
    count = Counter(data)
    # print(count)

    sorted_keys = sorted(count, key=count.get, reverse=True)
    # sorted_keys

    word_index = {word: i for i, word in enumerate(sorted_keys)}

  input_text = input("텍스트를 입력하세요. ")

  # 같은 조건으로 입력 텍스트 전처리
  input_text = input_text.lower()
  input_text = re.sub(r"[^가-힣a-zA-Z0-9' ]","",input_text)
  input_text = input_text.split()

  # 대응하는 요소의 정수 값을 딕셔너리에서 찾아 출력
  list_idx = []
  for idx in input_text:
    idx = word_index[idx]
    list_idx.append(idx)

  return list_idx

encoder()


```
